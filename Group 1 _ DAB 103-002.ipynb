# Question 3: Compare the range (capacity of distance traveled) between Plugin Hybrid Electric Vehieles and Battery Electric Vehicles.

```python
Battery_range =  Battery_Operated[['Make','Vehicle_Class','Range(km)']] 
```

```python
Battery_range=Battery_range.dropna(subset=['Vehicle_Class','Make'])
```

```python
Battery_range.info()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(Battery_range['Vehicle_Class'])
plt.title('Range of Battery Operated Cars')
plt.xlabel('Vehicle Class')
plt.xticks(rotation=90)
plt.ylabel('Range(km)')
```

```python
Hybrid_Range = Hybrid_Vehicles[['Make','Vehicle Class','Range 1', 'Range 2']]
```

```python
Hybrid_Range.rename(columns = {'Range 1': 'Range 1(km)','Range 2': 'Range 2(km)'}, inplace = True);Hybrid_Range
```

```python
Hybrid_Range = Hybrid_Range.drop (labels = 0, axis = 0,
           inplace = False)
```

```python
plt.scatter(Hybrid_Range['Vehicle Class'], Hybrid_Range['Range 1(km)'], marker='o', color='y',edgecolors= "black", s=100)
plt.scatter(Hybrid_Range['Vehicle Class'], Hybrid_Range['Range 2(km)'], marker='^', color='r',edgecolors= "black",s=100)
plt.xticks(rotation=90)
plt.plot()
plt.xlabel("Vehicle Class")
plt.ylabel("Range(km)")
plt.title("Range of Hybrid Cars(in Battery & Fuel)")
plt.legend(['Range in Electricity', 'Range in Fuel'],bbox_to_anchor=(1.3, 1.05), frameon = True);
plt.show()
```
