# Question 1: What is the difference between consumption of fuel in Gas Powered Vehicles, Plug-in Hybrid Electric Vehicles and Battery Electric Vehicles.?

The fuel consumption for each category is explained in graphical representatioin below. To get the comparison, the data shows the fuel consumption for gas powered and Plug-in Hybrid Electric Vehicles by converting the electricity utilization to fuel, which is ***One litre of gasoline contains the energy equivalent to 8.9 kWh of electricity***

By the results of visualization, the fuel usage is high in Hybrid vehicles in comparison to battery powered vehicles. This is due to the fact that electricity is consumed more in Hybrid vehicles and to get the difference between them, the elctricity is converted into fuel as explained above.

```python
Battery_Consumption = Battery_Operated.groupby(['Vehicle_Class'])['Fuel_Consumption_Comb(Le/100km)'].mean(); Battery_Consumption
```

```python
Battery_Consumption.plot(kind = 'bar',
                        title = 'Gas Consumption by Battery Operated Vehicles',
                        ylabel = '(Gas consumption for city and hwy combined(L/100km)',
                        xlabel = 'Vehicle Class',
               color='#3366FF')
```

The above graph represents the consumption of fuel in **Battery Operated** vehicles of different type. The **Pickup Trucks** and **Compact Vehicles** have the most usage of Fuel. The visulaization shows the consumption of fuel with taking the combination of consumption of fuel in city and on highway.

# Question 3- What is the range (capacity of distance traveled) between Plug-in Hybrid Electric Vehicles and Battery Electric Vehicles?

The visualization suggests that a vehicle's range changes based on the type of energy used. 
Hybrid vehicles range is greater compared to a battery-powered vehicle, as it has a longer range since they are powered by both, fuel and energy.
For instance, for the Mid-size car type when it’s hybrid running on fuel, it has a range capacity of 995 km compared to the other category.

```python
Battery_range =  Battery_Operated[['Make','Vehicle_Class','Range(km)']] 
```

```python
Battery_range=Battery_range.dropna(subset=['Vehicle_Class','Make'])
```

```python
Battery_range.info()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(Battery_range['Vehicle_Class'],palette= ["#99FF99"])
plt.title('Range of Battery Operated Cars',size=18,fontweight="bold")
plt.xlabel('Vehicle_Class')
plt.xticks(rotation=90)
plt.ylabel('Range(km)')
```

```python
Battery_subset2 = Battery_Operated[['Vehicle_Class','Range(km)']]
Range_subset1 = Hybrid_Vehicles[['Vehicle_Class','Range_1']]
Range_subset2 = Hybrid_Vehicles[['Vehicle_Class','Range_2']]
Battery_subset2.rename(columns = {'Range(km)': 'Range'}, inplace = True)
Battery_subset2
```

```python
Range_subset1
Range_subset2
```

```python
plt.bar(Range_subset2.Vehicle_Class, Range_subset2.Range_2, label="Hybrid_Fuel_Range (km)", color='#99FF99')
plt.bar(Battery_subset2.Vehicle_Class, Battery_subset2.Range, label ='Battery_powered_Range(km)', color='#CCCCCC')
plt.bar(Range_subset1.Vehicle_Class, Range_subset1.Range_1,label='Hybrid_Electric_Range (km)', color='#3366FF')
plt.xticks(rotation=90)
plt.plot()

plt.xlabel("Vehicle_class")
plt.ylabel("Range in km")
plt.title("Range distribution in Hybrid Cars & Battery-operated Cars",size=13,fontweight="bold")
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```
